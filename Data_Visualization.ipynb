```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("clean_house_price_dataset.csv")

# 1. Histogram of House Prices
plt.figure(figsize=(8,6))
sns.histplot(df['price in $'], bins=50, kde=True)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

```

```python

# 2. Boxplot of Prices by Region (replace 'region' with your column name)
plt.figure(figsize=(10,6))
sns.boxplot(x='state', y='price in $', data=df)
plt.title("House Prices by Region")
plt.xticks(rotation=45)
plt.show()
```

```python

# 3. Scatter Plot: Price vs Square Footage
plt.figure(figsize=(8,6))
sns.scatterplot(x='livingArea in sq.ft', y='price in $', data=df, alpha=0.5)
plt.title("Price vs Square Footage")
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.show()
```
