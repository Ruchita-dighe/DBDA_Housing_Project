```python
# ============================
# 1. Install & Import Libraries
# ============================
!pip install sdv pandas matplotlib seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ctgan import CTGAN


# ============================
# 2. Load Your Demo Dataset
# ============================
df = pd.read_csv(filepath + "house_price_dataset_cleaned.csv")

# ============================
# 3. Quick Profiling
# ============================
print(df.describe(include='all'))
```

```python
df.isnull().sum()
```

```python
discrete_columns = [
    'beds',
    'baths',
    'zipcode',
    'city',
    'state',
    'country',
    'address'
]

model = CTGAN(
    epochs=500,          # increase for better learning
    batch_size=128,      # adjust based on dataset size
    verbose=True,
    pac=1 # Set pac to 1 to avoid AssertionError when batch_size is not a multiple of pac
)

model.fit(df, discrete_columns=discrete_columns)
```

```python
# ============================
# 5. Generate Synthetic Data
# ============================
synthetic_df = model.sample(1000000)
synthetic_df.head()
```

```python
# ============================
# 6. Save Synthetic Dataset
# ============================
synthetic_df.to_csv(filepath + "zillow_large_dataset.csv", index=False)
print("Synthetic dataset saved as zillow_large_dataset.csv")
```
